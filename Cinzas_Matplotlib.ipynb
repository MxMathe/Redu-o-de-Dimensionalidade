```python
import numpy as np
from PIL import Image
from IPython.display import Image as IPImage, display

# Função para carregar a imagem
def load_image(image_path):
    return Image.open(image_path)
```

*Função para converter a imagem para níveis de cinza*

```python
def rgb_to_grayscale(image):
    image = np.array(image)
    # Fórmula para converter para níveis de cinza
    grayscale_image = 0.2989 * image[:,:,0] + 0.5870 * image[:,:,1] + 0.1140 * image[:,:,2]
    return grayscale_image.astype(np.uint8)
```

*Função para binarizar a imagem*

```python
def binarize_image(grayscale_image, threshold=128):
    # Binarização: 0 para pixels abaixo do limiar, 255 para pixels acima
    return np.where(grayscale_image >= threshold, 255, 0).astype(np.uint8)
```

*Carregar a imagem*

```python
def show_images(image, grayscale_image, binarized_image):
    # Salvar as imagens temporariamente
    image_pil = Image.fromarray(image)
    grayscale_pil = Image.fromarray(grayscale_image)
    binarized_pil = Image.fromarray(binarized_image)

    # Salvar as imagens como arquivos temporários
    image_pil.save("/content/imagem_original.jpg")
    grayscale_pil.save("/content/imagem_cinza.jpg")
    binarized_pil.save("/content/imagem_binarizada.jpg")

    # Exibir as imagens no Colab
    display(IPImage(filename="/content/imagem_original.jpg"))
    display(IPImage(filename="/content/imagem_cinza.jpg"))
    display(IPImage(filename="/content/imagem_binarizada.jpg"))
```

```python
image_path = "/content/Imagem/M3.jpg"  # Caminho da imagem no Colab
image = load_image(image_path)
```

*Converter para níveis de cinza*

```python
grayscale_image = rgb_to_grayscale(image)
```

*Binarizar a imagem*

```python
binarized_image = binarize_image(grayscale_image)
```

*Exibir as imagens*

```python
show_images(np.array(image), grayscale_image, binarized_image)
```
